```python
#Importing the requred packages 

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the dataset 

df=pd.read_csv('Admission_Predict.csv')
df
```

```python
#removing the Serial No. coloumn from the dataset 
# df=df.iloc[row_dimention,column_dimention]
#df=df.iloc[start:end,start:end]

df=df.iloc[:,1:]
df
```

```python
#Display the coloumns present in the dataset 

df.columns
```

```python
# add new Admission coloumn in the dataset 
# if chance of admit is greater than 0.75 then add this data into the chances of admit coloumn

df['Admission']=df["Chance of Admit "]>0.75
```

```python
# display the dataframe 
# we will get True and False values in the Admission Column 

df
```

```python
# We need data in olny interger formnat so comvert True and false with 0 & 1 
# df=df.replace({old : new , old : new})

df=df.replace({True:1,False:0})
df
```

```python
# drop Chance of Admit column from the dataset 

df=df.drop('Chance of Admit ',axis=1)
df
```

```python
#import the model_selection from sklearn library for using the train_test_split()
import sklearn.model_selection as tts

#import the tree from sklearn library for using the DecisionTreeClassifier()
import sklearn.tree as tree
```

```python
# split the dataset into x 
# take all columns excluding the Admission column

#df=df.iloc[row_dimention,column_dimention]
#df=df.iloc[start:end,start:end]
x=df.iloc[:,0:-1]
x
```

```python
# split the dataset into y
# take only the Admission column

# df=df.iloc[row_dimention,column_dimention]
#df=df.iloc[start:end,start:end]
y=df.iloc[:,-1:]
y
```

```python
#apply train_test_split on the dataset 

#train data of x, test data of x, train data of y, test data of y =  sklearn.model_selection.train_test_split(x,y,test_size=0.25)
train_x,test_x,train_y,test_y=tts.train_test_split(x,y,test_size=0.25)
```

```python
# Display trainnig data of the x dataset 
#300 rows × 6 columns

train_x
```

```python
# Display tresting data of the x dataset 
#100 rows × 6 columns

test_x
```

```python
# Display trainnig data of the y dataset 
#300 rows × 6 columns

train_y
```

```python
# Display testing data of the y dataset 
#100 rows × 6 columns 

test_y
```

```python
#create instance of the sklearn.tree.DecisionTreeClassifier()

#variable name = sklearn.tree.DecisionTreeClassifier(criterion='gini')
#criterion='gini' is by default provided by library

gini_decision_tree_model=tree.DecisionTreeClassifier()
```

```python
# Fit the training data into the model 

#instance of the  DecisionTreeClassifier(). fit (training data of x & y)
gini_decision_tree_model.fit(train_x,train_y)
```

```python
# calculate the score of the trained model 
#instance of the  DecisionTreeClassifier().score(testing data of x & y)

gini_decision_tree_model.score(test_x,test_y)
```

```python
# Plotting the trained decision tree 
# saving the PDF file of the decsion tree

tree.plot_tree(gini_decision_tree_model)
plt.savefig('gini_decision_tree_model.pdf')
plt.show()
```

```python
#create instance of the sklearn.tree.DecisionTreeClassifier()

#variable name = sklearn.tree.DecisionTreeClassifier(criterion='entropy'')
entropy_decision_tree_model=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
# Fit the training data into the model 

#instance of the  DecisionTreeClassifier(). fit (training data of x & y)
entropy_decision_tree_model.fit(train_x,train_y)
```

```python
# calculate the score of the trained model 
#instance of the  DecisionTreeClassifier().score(testing data of x & y)

entropy_decision_tree_model.score(test_x,test_y)
```

```python
# Plotting the trained decision tree 
# saving the PDF file of the decsion tree

tree.plot_tree(entropy_decision_tree_model)
plt.savefig("entropy_decision_tree_model.pdf")
plt.show()
```
